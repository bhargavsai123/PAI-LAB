```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import ParameterEstimator
from pgmpy.inference import VariableElimination
import pandas as pd

data = pd.DataFrame(data={'Rain' : ['No','No','Yes','Yes','No','Yes','Yes','No'],
                    'TrafficJam' : ['Yes','No','Yes','No','Yes','Yes','No','No'],
                    'ArriveLate' : ['Yes','No','Yes','No','No','Yes','Yes','No']})


model = DiscreteBayesianNetwork([('Rain','TrafficJam'),('TrafficJam','ArriveLate')])

model.fit(data)

print(model.get_cpds())

inference = VariableElimination(model)

query_result = inference.query(variables=['ArriveLate'],evidence={'Rain':'Yes'})
print(query_result)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

model = DiscreteBayesianNetwork([('Income','Default'),('Credit','Default')])

cpd_income = TabularCPD('Income',2,[[0.6],[0.4]])
cpd_credit = TabularCPD('Credit',2,[[0.7],[0.3]])
cpd_default = TabularCPD('Default',2,[[0.95,0.9,0.8,0.5],[0.05,0.1,0.2,0.5]],evidence=['Income','Credit'],evidence_card=[2,2])

model.add_cpds(cpd_income,cpd_credit,cpd_default)

infer = VariableElimination(model)
query_result = infer.query(['Default'], evidence={'Income':1,'Credit':0})
print(query_result)
```

```python
import pandas as pd
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.inference import VariableElimination

data = pd.DataFrame({
    "Disease": ["Yes","Yes","Yes","Yes","Yes","No","No","No","No","No","Yes","No"],
    "Fever": ["Yes","Yes","No","Yes","No","Yes","Yes","No","No","Yes","Yes","No"],
    "Cough": ["Yes","No","Yes","Yes","No","Yes","No","Yes","No","Yes","Yes","No"],
    "Test_Result": ["Positive","Positive","Positive","Positive","Negative",
                    "Negative","Negative","Negative","Negative","Positive","Positive",
                    "Negative"]
})

model = DiscreteBayesianNetwork([('Disease','Fever'),('Disease','Cough'),('Fever','Test_Result')])

model.fit(data)

print(model.get_cpds())

inference = VariableElimination(model)

query_result = inference.query(variables=['Test_Result'],evidence={'Cough':'Yes'})
print(query_result)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
import pandas as pd

model = DiscreteBayesianNetwork([('Rain','Sprinkler'),('Rain','GrassWet'),('Sprinkler','GrassWet')])

cpd_rain = TabularCPD('Rain',2,[[0.8],[0.2]])
cpd_sprinkler = TabularCPD('Sprinkler',2,[[0.6,0.99],[0.4,0.01]],evidence=['Rain'],evidence_card=[2])
cpd_grasswet = TabularCPD('GrassWet',2,[[1,0.2,0.1,0.01],[0,0.8,0.9,0.99]],evidence=['Sprinkler','Rain'],evidence_card=[2,2])

model.add_cpds(cpd_rain,cpd_sprinkler,cpd_grasswet)

infer = VariableElimination(model)
query = infer.query(['Rain'],evidence={'GrassWet':1})
print(query)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
import pandas as pd
model=DiscreteBayesianNetwork([('B','A'),('T','A'),('E','A'),('E','T')])
cpt_b=TabularCPD(variable='B',variable_card=2,values=[[0.7],[0.3]])
cpt_e=TabularCPD(variable='E',variable_card=2,values=[[0.4],[0.6]])
cpt_t=TabularCPD(variable='T',variable_card=2,values=[[0.8,0.5],[0.2,0.5]],evidence=['E'],evidence_card=[2])
cpt_a=TabularCPD(variable='A',variable_card=2,values=[[1,0.9,0.95,0.85,0.89,0.7,0.87,0.3],[0,0.1,0.05,0.15,0.11,0.3,0.13,0.7]],evidence=['E','B','T'],evidence_card=[2,2,2])
model.add_cpds(cpt_b,cpt_e,cpt_a,cpt_t)
infer=VariableElimination(model)
result=infer.query(variables=['E'],evidence={'A':0})
print(result)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
import pandas as pd

model = DiscreteBayesianNetwork([('E','S'),('M','S'),('M','B')])
cpd_e = TabularCPD('E',2,[[0.6],[0.4]])
cpd_m = TabularCPD('M',2,[[0.9],[0.1]])
cpd_b = TabularCPD('B',2,[[0.9,0.0],[0.1,1.0]],evidence=['M'],evidence_card=[2])
cpd_s = TabularCPD('S',2,[[0.9,0.7,0.2,0.0],[0.1,0.3,0.8,1.0]],evidence=['E','M'],evidence_card=[2,2])
model.add_cpds(cpd_e,cpd_m,cpd_b,cpd_s)
infer=VariableElimination(model)
result=infer.query(variables=['E','M','S','B'])
print("1. P(−E,−S,−M,−B) : ",result.get_value(E=0,S=0,M=0,B=0))
print("2. P(B) : ",infer.query(variables=['B']).get_value(B=1))
print("3. P(M/B) : ",infer.query(variables=['M'],evidence={'B':1}).get_value(M=1))
print("4. P(M/S,B,E) : ",infer.query(variables=['M'],evidence={'S':1,'B':1,'E':1}).get_value(M=1))
print("5. P(E/M) : ",infer.query(variables=['E'],evidence={'M':1}).get_value(E=1))
```
